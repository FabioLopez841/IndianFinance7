```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

data = pd.read_csv('data.csv')
```

```python
# Pregunta 1: Â¿CuÃ¡l es la distribuciÃ³n de los ingresos en el conjunto de datos?
plt.figure(figsize=(10, 6))
sns.histplot(data['Income'], bins=30, kde=True)
plt.title('DistribuciÃ³n de Ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.show()
```

 La distribuciÃ³n de los ingresos muestra cÃ³mo se distribuyen los valores de ingreso en el conjunto de datos.

```python
# Pregunta 2: Â¿CÃ³mo varÃ­an los ingresos segÃºn la edad de los individuos?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Income')
plt.title('RelaciÃ³n entre Edad e Ingresos')
plt.xlabel('Edad')
plt.ylabel('Ingresos')
plt.show()

# Markdown: 
# Este grÃ¡fico de dispersiÃ³n ilustra la relaciÃ³n entre la edad y los ingresos, permitiendo observar tendencias.
```

```python
# Pregunta 3: Â¿CuÃ¡l es la relaciÃ³n entre el nÃºmero de dependientes y los ingresos?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Dependents', y='Income')
plt.title('Ingresos por NÃºmero de Dependientes')
plt.xlabel('NÃºmero de Dependientes')
plt.ylabel('Ingresos')
plt.show()

# Markdown: 
# El boxplot muestra cÃ³mo varÃ­an los ingresos en funciÃ³n del nÃºmero de dependientes.
```

```python
# Pregunta 4: Â¿QuÃ© ocupaciones tienen los ingresos mÃ¡s altos?
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Occupation', y='Income', estimator='mean')
plt.title('Media de Ingresos por OcupaciÃ³n')
plt.xlabel('OcupaciÃ³n')
plt.ylabel('Ingresos Promedio')
plt.xticks(rotation=45)
plt.show()

# Markdown: 
# Este grÃ¡fico de barras muestra la media de ingresos para cada ocupaciÃ³n, destacando las mÃ¡s lucrativas.
```

```python
# Pregunta 5: Â¿CÃ³mo se distribuyen los gastos en comestibles por ciudad?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='City_Tier', y='Groceries')
plt.title('Gastos en Comestibles por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Gastos en Comestibles')
plt.show()

# Markdown: 
# El boxplot ilustra la variabilidad de los gastos en comestibles segÃºn la ciudad.
```

```python
# Pregunta 6: Â¿CuÃ¡l es la relaciÃ³n entre el ingreso disponible y los ahorros deseados?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Disposable_Income', y='Desired_Savings')
plt.title('RelaciÃ³n entre Ingreso Disponible y Ahorros Deseados')
plt.xlabel('Ingreso Disponible')
plt.ylabel('Ahorros Deseados')
plt.show()

# Markdown: 
# Este grÃ¡fico de dispersiÃ³n muestra la relaciÃ³n entre el ingreso disponible y los ahorros deseados.
```

```python
# Pregunta 7: Â¿QuÃ© porcentaje del ingreso se destina a diferentes categorÃ­as de gastos?
# Crear un DataFrame para los gastos
gastos = data[['Transport', 'Eating_Out', 'Entertainment', 'Utilities', 'Healthcare', 'Education', 'Miscellaneous']]
gastos_sum = gastos.sum()

plt.figure(figsize=(10, 6))
gastos_sum.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('ProporciÃ³n de Gastos por CategorÃ­a')
plt.ylabel('')
plt.show()

# Markdown: 
# El grÃ¡fico de pastel muestra la proporciÃ³n de gastos en diferentes categorÃ­as.
```

```python
# Pregunta 8: Â¿CÃ³mo varÃ­an los gastos en entretenimiento segÃºn la ciudad?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='City_Tier', y='Entertainment')
plt.title('Gastos en Entretenimiento por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Gastos en Entretenimiento')
plt.show()

# Markdown: 
# Este boxplot muestra la variabilidad de los gastos en entretenimiento segÃºn la ciudad.
```

```python
# Pregunta 10: Â¿QuÃ© tan efectivos son los ahorros potenciales en diferentes categorÃ­as?
plt.figure(figsize=(10, 6))
potential_savings = data[['Potential_Savings_Groceries', 'Potential_Savings_Transport', 
                           'Potential_Savings_Eating_Out', 'Potential_Savings_Entertainment', 
                           'Potential_Savings_Utilities', 'Potential_Savings_Healthcare', 
                           'Potential_Savings_Education', 'Potential_Savings_Miscellaneous']]
potential_savings_sum = potential_savings.sum()

plt.figure(figsize=(10, 6))
potential_savings_sum.plot.bar()
plt.title('Ahorros Potenciales por CategorÃ­a')
plt.xlabel('CategorÃ­a')
plt.ylabel('Ahorros Potenciales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pregunta 11: Â¿CÃ³mo varÃ­an los gastos en salud segÃºn la ocupaciÃ³n?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Occupation', y='Healthcare')
plt.title('Gastos en Salud por OcupaciÃ³n')
plt.xlabel('OcupaciÃ³n')
plt.ylabel('Gastos en Salud')
plt.xticks(rotation=45)
plt.show()

# Markdown: 
# El boxplot ilustra cÃ³mo varÃ­an los gastos en salud segÃºn la ocupaciÃ³n, destacando diferencias significativas.
```

```python
# Pregunta 12: Â¿CuÃ¡l es la tendencia de los gastos en utilidades a lo largo de diferentes niveles de ingreso?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Income', y='Utilities')
plt.title('RelaciÃ³n entre Ingreso y Gastos en Utilidades')
plt.xlabel('Ingreso')
plt.ylabel('Gastos en Utilidades')
plt.show()

# Markdown: 
# Este grÃ¡fico de dispersiÃ³n muestra la relaciÃ³n entre el ingreso y los gastos en utilidades, permitiendo observar tendencias.
```

```python
# Pregunta 13: Â¿QuÃ© tan comunes son los diferentes niveles de ahorro deseado?
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Desired_Savings_Percentage')
plt.title('Frecuencia de Porcentaje de Ahorros Deseados')
plt.xlabel('Porcentaje de Ahorros Deseados')
plt.ylabel('Frecuencia')
plt.show()

# Markdown: 
# El grÃ¡fico de barras muestra la frecuencia de los diferentes niveles de ahorro deseado, ayudando a identificar patrones.
```

```python

# Pregunta 14: Â¿CÃ³mo se distribuyen los gastos en transporte segÃºn la edad?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Age', y='Transport')
plt.title('Gastos en Transporte por Edad')
plt.xlabel('Edad')
plt.ylabel('Gastos en Transporte')
plt.show()
```

```python
# Pregunta 15: Â¿CuÃ¡l es la relaciÃ³n entre el nivel de educaciÃ³n y los ingresos?
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Education', y='Income', estimator='mean')
plt.title('Media de Ingresos por Nivel de EducaciÃ³n')
plt.xlabel('Nivel de EducaciÃ³n')
plt.ylabel('Ingresos Promedio')
plt.xticks(rotation=45)
plt.show()

# Markdown: 
# Este grÃ¡fico de barras muestra la media de ingresos por nivel de educaciÃ³n, destacando la importancia de la educaciÃ³n en los ingresos.
```
