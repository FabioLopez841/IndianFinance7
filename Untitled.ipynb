```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

data = pd.read_csv('data.csv')
```

```python
data.columns
```

```python
data.describe()
```

```python
# Pregunta 1: ¿Cuál es la distribución de los ingresos en el conjunto de datos?
plt.figure(figsize=(10, 6))
sns.histplot(data['Income'], bins=30, kde=True)
plt.title('Distribución de Ingresos')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.show()
```

 La distribución de los ingresos muestra cómo se distribuyen los valores de ingreso en el conjunto de datos.

```python
# Pregunta 2: ¿Cómo varían los ingresos según la edad de los individuos?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Income')
plt.title('Relación entre Edad e Ingresos')
plt.xlabel('Edad')
plt.ylabel('Ingresos')
plt.show()

# Markdown: 
# Este gráfico de dispersión ilustra la relación entre la edad y los ingresos, permitiendo observar tendencias.
```

```python
# Pregunta 3: ¿Cuál es la relación entre el número de dependientes y los ingresos?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Dependents', y='Income')
plt.title('Ingresos por Número de Dependientes')
plt.xlabel('Número de Dependientes')
plt.ylabel('Ingresos')
plt.show()

# Markdown: 
# El boxplot muestra cómo varían los ingresos en función del número de dependientes.
```

```python
# Pregunta 4: ¿Qué ocupaciones tienen los ingresos más altos?
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Occupation', y='Income', estimator='mean')
plt.title('Media de Ingresos por Ocupación')
plt.xlabel('Ocupación')
plt.ylabel('Ingresos Promedio')
plt.xticks(rotation=45)
plt.show()

# Markdown: 
# Este gráfico de barras muestra la media de ingresos para cada ocupación, destacando las más lucrativas.
```

```python
# Pregunta 5: ¿Cómo se distribuyen los gastos en comestibles por ciudad?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='City_Tier', y='Groceries')
plt.title('Gastos en Comestibles por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Gastos en Comestibles')
plt.show()

# Markdown: 
# El boxplot ilustra la variabilidad de los gastos en comestibles según la ciudad.
```

```python
# Pregunta 6: ¿Cuál es la relación entre el ingreso disponible y los ahorros deseados?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Disposable_Income', y='Desired_Savings')
plt.title('Relación entre Ingreso Disponible y Ahorros Deseados')
plt.xlabel('Ingreso Disponible')
plt.ylabel('Ahorros Deseados')
plt.show()

# Markdown: 
# Este gráfico de dispersión muestra la relación entre el ingreso disponible y los ahorros deseados.
```

```python
# Pregunta 7: ¿Qué porcentaje del ingreso se destina a diferentes categorías de gastos?
# Crear un DataFrame para los gastos
gastos = data[['Transport', 'Eating_Out', 'Entertainment', 'Utilities', 'Healthcare', 'Education', 'Miscellaneous']]
gastos_sum = gastos.sum()

plt.figure(figsize=(10, 6))
gastos_sum.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Proporción de Gastos por Categoría')
plt.ylabel('')
plt.show()

# Markdown: 
# El gráfico de pastel muestra la proporción de gastos en diferentes categorías.
```

```python
# Pregunta 8: ¿Cómo varían los gastos en entretenimiento según la ciudad?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='City_Tier', y='Entertainment')
plt.title('Gastos en Entretenimiento por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Gastos en Entretenimiento')
plt.show()

# Markdown: 
# Este boxplot muestra la variabilidad de los gastos en entretenimiento según la ciudad.
```

```python
# Pregunta 10: ¿Qué tan efectivos son los ahorros potenciales en diferentes categorías?
plt.figure(figsize=(10, 6))
potential_savings = data[['Potential_Savings_Groceries', 'Potential_Savings_Transport', 
                           'Potential_Savings_Eating_Out', 'Potential_Savings_Entertainment', 
                           'Potential_Savings_Utilities', 'Potential_Savings_Healthcare', 
                           'Potential_Savings_Education', 'Potential_Savings_Miscellaneous']]
potential_savings_sum = potential_savings.sum()

plt.figure(figsize=(10, 6))
potential_savings_sum.plot.bar()
plt.title('Ahorros Potenciales por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ahorros Potenciales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pregunta 11: ¿Cómo varían los gastos en salud según la ocupación?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Occupation', y='Healthcare')
plt.title('Gastos en Salud por Ocupación')
plt.xlabel('Ocupación')
plt.ylabel('Gastos en Salud')
plt.xticks(rotation=45)
plt.show()

# Markdown: 
# El boxplot ilustra cómo varían los gastos en salud según la ocupación, destacando diferencias significativas.
```

```python
# Pregunta 12: ¿Cuál es la tendencia de los gastos en utilidades a lo largo de diferentes niveles de ingreso?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Income', y='Utilities')
plt.title('Relación entre Ingreso y Gastos en Utilidades')
plt.xlabel('Ingreso')
plt.ylabel('Gastos en Utilidades')
plt.show()

# Markdown: 
# Este gráfico de dispersión muestra la relación entre el ingreso y los gastos en utilidades, permitiendo observar tendencias.
```

```python
# Pregunta 13: ¿Qué tan comunes son los diferentes niveles de ahorro deseado?
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Desired_Savings_Percentage')
plt.title('Frecuencia de Porcentaje de Ahorros Deseados')
plt.xlabel('Porcentaje de Ahorros Deseados')
plt.ylabel('Frecuencia')
plt.show()

# Markdown: 
# El gráfico de barras muestra la frecuencia de los diferentes niveles de ahorro deseado, ayudando a identificar patrones.
```

```python

# Pregunta 14: ¿Cómo se distribuyen los gastos en transporte según la edad?
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Age', y='Transport')
plt.title('Gastos en Transporte por Edad')
plt.xlabel('Edad')
plt.ylabel('Gastos en Transporte')
plt.show()
```

```python
# Pregunta 15: ¿Cuál es la relación entre el nivel de educación y los ingresos?
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Education', y='Income', estimator='mean')
plt.title('Media de Ingresos por Nivel de Educación')
plt.xlabel('Nivel de Educación')
plt.ylabel('Ingresos Promedio')
plt.xticks(rotation=45)
plt.show()

# Markdown: 
# Este gráfico de barras muestra la media de ingresos por nivel de educación, destacando la importancia de la educación en los ingresos.
```
